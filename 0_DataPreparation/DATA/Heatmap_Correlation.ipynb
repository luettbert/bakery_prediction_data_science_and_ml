```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def create_correlation_heatmap(df, figsize=(10, 8)):
    """
    Erstellt eine Korrelations-Heatmap für einen DataFrame

    Parameters:
    df (pandas.DataFrame): DataFrame mit numerischen Spalten
    figsize (tuple): Größe der Abbildung (Breite, Höhe)
    """
    # Korrelationsmatrix berechnen
    corr_matrix = df.corr()

    # Figure erstellen
    plt.figure(figsize=figsize)

    # Heatmap erstellen
    sns.heatmap(
        corr_matrix,
        annot=True,  # Zeigt die Korrelationswerte an
        cmap='coolwarm',  # Farbschema: rot (positiv) bis blau (negativ)
        vmin=-1, vmax=1,  # Wertebereich der Korrelation
        center=0,  # Zentrum der Farbskala
        fmt='.2f',  # Zahlenformat (2 Dezimalstellen)
        square=True,  # Quadratische Zellen
    )

    # Titel hinzufügen
    plt.title('Feature Korrelations-Heatmap')

    # Layout optimieren
    plt.tight_layout()

    return plt

# Beispiel für die Verwendung
if __name__ == "__main__":
    # Beispieldaten erstellen
    data = {
        'Feature1': [1, 2, 3, 4, 5],
        'Feature2': [2, 4, 5, 4, 5],
        'Feature3': [3, 3, 3, 3, 3],
        'Feature4': [1, 3, 5, 7, 9]
    }
    df = pd.DataFrame(data)

    # Heatmap erstellen
    plt = create_correlation_heatmap(df)

    # Anzeigen der Heatmap
    plt.show()
```
